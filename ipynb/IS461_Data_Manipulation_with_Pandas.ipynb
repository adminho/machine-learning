### สร้างจาก dictionary <font color="red">(เสริมเพิ่มเติม)</font>

```
data = {
 'calories': [420, 380, 390],
 'duration': [50, 40, 45],
 'pulse':[90, 100, 109]
}
pd.DataFrame(data)
```

ตอนสร้าง เลือกคอลัมน์ที่ต้องการจาก dictionary ต้นฉบับก็ได้  (ไม่ต้องเอาคีย์มาทั้งหมด)

```
data = {
 'calories': [420, 380, 390],
 'duration': [50, 40, 45],
 'pulse': [90, 100, 109]
}

pd.DataFrame(data, columns = ['calories', 'pulse'])
```

สร้าง index เองได้

```
data = {
 'calories': [420, 380, 390],
 'duration': [50, 40, 45],
 'pulse': [90, 100, 109]
}

pd.DataFrame(data, index = ["item1", "item2", "item3"])
```

สร้างจาก dictionary ของ Series

```
data = {'one': pd.Series([10, 20, 30, 40],
                      index=['a', 'b', 'c', 'd']),
        'two': pd.Series([10, 20, 30, 40],
                      index=['a', 'b', 'c', 'd'])}

pd.DataFrame(data)
```

### สร้างจาก list ของ dictionary <font color="red">(เสริมเพิ่มเติม)</font>

ตั้งชื่อ index เองก็ได้

```
data = [{'calories':420, 'duration': 380, 'pulse': 390},
        {'calories': 50, 'duration': 40, 'pulse': 45},
        {'calories': 90, 'duration': 100, 'pulse': 109}]

pd.DataFrame(data, index = ["item1", "item2", "item3"])
```

สามารถเลือกคอลัมน์ตามที่ต้องการได้

```
data = [{'calories':420, 'duration': 380, 'pulse': 390},
        {'calories': 50, 'duration': 40, 'pulse': 45},
        {'calories': 90, 'duration': 100, 'pulse': 109}]

pd.DataFrame(data,  index = ["item1", "item2", "item3"], columns = ['calories', 'pulse'])
```

### สร้างจาก list ของ tuple <font color="red">(เสริมเพิ่มเติม)</font>

```
data =[('Boy', 25),
       ('Robert', 30),
       ('Pan', 26),
        ('Sak', 22)]
pd.DataFrame(data)
```

สามารถระบุชื่อคอลัมน์เองได้

```
data =[('Boy', 25),
       ('Robert', 30),
       ('Pan', 26),
        ('Sak', 22)]
pd.DataFrame(data, columns=['name', 'age'])
```

หรือจะใช้ฟังก์ชั่น zip() สร้าง tuple ขึ้นมาก่อน

```
# List1
Name = ['Boy', 'Robert', 'Pan', 'Sak']

# List2
Age = [25, 30, 26, 22]


list_of_tuples = list(zip(Name, Age))
print("Show tuple: ", list_of_tuples)


pd.DataFrame(list_of_tuples,
                  columns=['name', 'age'])
```

# แอตทริบิวต์ index , column, values ของ DataFrame



**อินเดกซ์**

แอตทริบิวต์ df.index จะเป็นชนิดข้อมูล RangeIndex มีไว้เก็บข้อมูล index ของ df นั้นทั้งหมด

  * ตรงตำแหน่ง cells ที่ว่าง ก็เติมให้ครบ
  * อาจแทนที่ด้วย Mean, Median, หรือ Mode
* รูปแบบข้อมูลผิด ("Date" ที่ row 26) ให้แก้ไข เช่น

* ข้อมูลผิด ("Duration" ที่ row 7) ให้แก้ไข เช่น
  * ลบแถวนั้น
  * แก้ให้ถูก
* ข้อมูลซ้ำ (row 11 กับ 12) ให้แก้ไข
   * ลบแถวที่ซ้ำ

**ลบแถว ที่มี cells ไม่มีข้อมูล (์Null)**

ใช้ฟังก์ชัน dropna()
ลบค่าพวก NaN (Not a Number), NaT (Not a Time) เป็นต้น

ถ้าไม่ระบุ inplace=True ก็จะไม่แก้ไขข้อมูลต้นฉบับ แต่จะรีเทิร์น dataframe ตัวใหม่ตัวใหม่ที่ถูกแก้ไข

```
import pandas as pd
df = pd.read_csv('https://www.w3schools.com/python/pandas/dirtydata.csv')

df.dropna(inplace=True) # แถวที่ 22, 18, 28 ถูกลบไปแล้ว
df[17:30]
```

หรือจะลบบาง column ที่มี NaN

```
import pandas as pd
df = pd.read_csv('https://www.w3schools.com/python/pandas/dirtydata.csv')
df.dropna(subset=['Calories'], inplace = True) #  # แถวที่ 22 ซึ่งมี NaN ยังอยู่ แต่แถวที่ 18, 28 ถูกลบไปแล้ว
df[17:30]
```

**เติมเต็มแถวที่มี Null ด้วยข้อมูลอื่น (เติมทั้งแถว)**

ใช้ฟังก์ชัน fillna() เติมค่าที่เป็น Null

ถ้าไม่ระบุ inplace=True ก็จะไม่แก้ไขข้อมูลต้นฉบับ แต่จะรีเทิร์น dataframe ตัวใหม่ที่ถูกแก้ไข

```
import pandas as pd
df = pd.read_csv('https://www.w3schools.com/python/pandas/dirtydata.csv')

df.fillna(130 , inplace = True) # เติมแถวที่ 18, 22, 28
df.loc[[18, 22, 28]]
```

หรือเติมเต็มเฉพาะคอลัมน์ที่กำหนด

```
import pandas as pd
df = pd.read_csv('https://www.w3schools.com/python/pandas/dirtydata.csv')

df['Calories'].fillna(130, inplace = True)  # เติมแถวที่ 18, 28 ส่วนแถวที่ 22 ยังไม่ได้เติม
df.loc[[18, 22, 28]]
```

**หาค่า Mean, Median, or Mode ทั้งคอลัมน์**
* ฟังก์ชัน mean() หาค่าเฉลี่ย
* ฟังก์ชัน median() หาค่ามัธยฐาน
* ฟังก์ชั่น mode() ค่าจำนวนที่พบบ่อยมากสุด

```
import pandas as pd
df = pd.read_csv('https://www.w3schools.com/python/pandas/dirtydata.csv')

mean = df["Calories"].mean()
df["Calories"].fillna(mean, inplace =True)
df.loc[[18, 28]]
```

```
import pandas as pd
df = pd.read_csv('https://www.w3schools.com/python/pandas/dirtydata.csv')

median = df["Calories"].median()
df["Calories"].fillna(median, inplace =True)
df.loc[[18, 28]]
```

```
import pandas as pd
df = pd.read_csv('https://www.w3schools.com/python/pandas/dirtydata.csv')

mode = df["Calories"].mode()[0]
df["Calories"].fillna(mode, inplace =True)
df.loc[[18, 28]]
```

สามารถแปลงformat date ทั้งคอลัมน์

ถ้าแปลงเป็น date ไม่ได้ ก็จะได้ NaT (Not a Time) ซึ่งถือว่าเป็น Null
สามารถใช้ dropna() ลบทิ้งได้ หรือเติมค่าใหม่ด้วย fillna

```
import pandas as pd
df = pd.read_csv('https://www.w3schools.com/python/pandas/dirtydata.csv')
df['Date'] = pd.to_datetime(df['Date'])

df.dropna(subset=['Date'], inplace = True) # ลบแถวที่ 22 ทิ้ง เพราะมี NaT (มาจาก pd.to_datetime())
df[20:25]
```

ลบแถวที่ซ้ำด้วยฟังก์ชัน drop_duplicates
ถ้าไม่ระบุ inplace=True ก็จะไม่แก้ไขข้อมูลต้นฉบับ แต่จะรีเทิร์น dataframe ตัวใหม่ที่ถูกแก้ไข

```
import pandas as pd
df = pd.read_csv('https://www.w3schools.com/python/pandas/dirtydata.csv')

df.drop_duplicates(inplace = True)  # ลบแถวที่ 12 ที่ซ้ำซ้อนทิ้งไป
df[10:13]
```

ฟังก์ชัน duplicated()
จะค้นหาแถวที่ซ้ำ
 โดยจะรีเทิร์นเป็น series ที่สมาชิกภายในจะแสดงค่า True บอกว่าซ้ำ ถ้าไม่ซ้ำเป็น False

```
import pandas as pd
df = pd.read_csv('https://www.w3schools.com/python/pandas/dirtydata.csv')
found = df.duplicated()
found[5:15]
```
