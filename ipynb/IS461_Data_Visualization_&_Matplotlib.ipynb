## พล็อตกราฟหลายเส้น

พล็อตกราฟหลายเส้นอีกวิธีหนึ่ง สามารถเรียก plt.plot() แล้วระบุชุด (x, y) ต่อๆ กันก็ได้

```
#https://pythonspot.com/matplotlib-line-chart/
#from pylab import *
import matplotlib.pyplot as plt
import numpy as np
import math

plt.xlabel('time (s)')
plt.ylabel('voltage (mV)')
plt.title('Sine Wave')
plt.ylim(-2,2)

t = np.arange(0.0, 2.0, 0.01)
s1 = np.sin(2.5*math.pi*t)
plt.plot(t, s1)

s2 = np.cos(2.5*math.pi*t)
plt.plot(t, s2)

plt.show()
```

```
#https://pythonspot.com/matplotlib-line-chart/
#from pylab import *
import matplotlib.pyplot as plt
import numpy as np
import math

plt.xlabel('time (s)')
plt.ylabel('voltage (mV)')
plt.title('Sine Wave')
plt.ylim(-2,2)

t = np.arange(0.0, 2.0, 0.01)
s1 = np.sin(2.5*math.pi*t)
s2 = np.cos(2.5*math.pi*t)

plt.plot(t, s1)

plt.show()
```

เมื่อพล็อตกราฟแล้ว ปกติจะได้ข้อมูลกราฟ มาเก็บไว้

```
#https://pythonspot.com/matplotlib-line-chart/
#from pylab import *
import matplotlib.pyplot as plt
import numpy as np
import math


plt.xlabel('time (s)')
plt.ylabel('voltage (mV)')
plt.title('Sine Wave')
plt.ylim(-2,2)

t = np.arange(0.0, 2.0, 0.01)
s1 = np.sin(2.5*math.pi*t)
s2 = np.cos(2.5*math.pi*t)

#return list of Line2D
line = plt.plot(t, s1, t, s2)

plt.show()

print("Detail of each line.")
#print( line[0].get_data() )
#print( line[1].get_data() )
print( line[0].get_xdata() )
print( line[1].get_xdata() )
```

เปลี่ยนสีเส้น เปลี่ยนขนาดเส้น เปลี่ยนไสตล์เส้น ให้เห็นถึงความแตกต่าง

```
#https://pythonspot.com/matplotlib-line-chart/
#from pylab import *
import matplotlib.pyplot as plt
import numpy as np
import math

plt.xlabel('time (s)')
plt.ylabel('voltage (mV)')
plt.title('Sine Wave')
plt.ylim(-2,2)

t = np.arange(0.0, 2.0, 0.01)
s1 = np.sin(2.5*math.pi*t)
plt.plot(t, s1, 'm-')

t = np.arange(0.0, 2.0, 0.01)
s2 = np.cos(2.5*math.pi*t)
plt.plot(t, s2, 'g--')

#plt.legend(['cos', 'sine']) # ระบุ label ภายหลังก็ได้

plt.show()
```

## กำหนด legend <

* สามารถติด label ตรงเส้น
* เพิ่ม legend

อ่านเพิ่มเติม https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html

```
#https://pythonspot.com/matplotlib-line-chart/
#from pylab import *
import matplotlib.pyplot as plt
import numpy as np
import math

plt.xlabel('time (s)')
plt.ylabel('voltage (mV)')
plt.title('Sine Wave')
plt.ylim(-2,2)

t = np.arange(0.0, 2.0, 0.01)
s1 = np.sin(2.5*math.pi*t)
plt.plot(t, s1, 'm-', label='sin(t)')

t = np.arange(0.0, 2.0, 0.01)
s2 = np.cos(2.5*math.pi*t)
plt.plot(t, s2, 'g--', label='cos(t)')

plt.legend()
#plt.legend(['cos', 'sine']) # ระบุ label ภายหลังก็ได้

plt.show()
```

ติด labels ทีหลังก็ได้

```
#https://pythonspot.com/matplotlib-line-chart/
#from pylab import *
import matplotlib.pyplot as plt
import numpy as np
import math

plt.xlabel('time (s)')
plt.ylabel('voltage (mV)')
plt.title('Sine Wave')
plt.ylim(-2,2)

t = np.arange(0.0, 2.0, 0.01)
s2 = np.sin(2.5*math.pi*t)
#plt.plot(t, s2, 'm-', label='cos(x)')
plt.plot(t, s2, 'm-')

t = np.arange(0.0, 2.0, 0.01)
s2 = np.cos(2.5*math.pi*t)
#plt.plot(t, s2, 'g--', label='sin(x)')
plt.plot(t, s2, 'g--')

#plt.legend()
plt.legend(['cos', 'sine']) # ระบุ label ภายหลังก็ได้

plt.show()
```

ปรับแต่ง legend

```
#https://pythonspot.com/matplotlib-line-chart/
#from pylab import *
import matplotlib.pyplot as plt
import numpy as np
import math

plt.xlabel('time (s)')
plt.ylabel('voltage (mV)')
plt.title('Sine Wave')
plt.ylim(-2,2)

t = np.arange(0.0, 2.0, 0.01)
s1 = np.sin(2.5*math.pi*t)
plt.plot(t, s1, 'm-', label='sin(t)')

t = np.arange(0.0, 2.0, 0.01)
s2 = np.cos(2.5*math.pi*t)
plt.plot(t, s2, 'g--', label='cos(t)')

#plt.legend()
#plt.legend(['cos', 'sine']) # ระบุ label ภายหลังก็ได้

# framealpha ถ้าเข้าใกล้ 0 จะโปรงใสจนเห็นเงาดำด้านหลัง ถ้าเป็น 1 จะทึบแสงจนมองไม่เห็นเงาด้านหลัง
plt.legend(loc='upper center', ncol=2, frameon=True, fancybox=True, framealpha=1, shadow=True, borderpad=1)

plt.show()
```

```
#https://pythonspot.com/matplotlib-bar-chart/

import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt

lang = ['Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp']
usage = [10, 8, 6, 4, 2, 1]

plt.bar(lang, usage, align='center', alpha=0.5)

#plt.xticks(y_pos, objects)
plt.ylabel('Usage')
plt.title('Programming language usage')

plt.show()
```

## Pie chart

อ่านเพิ่ม
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html

```
#  https://pythonspot.com/matplotlib-pie-chart/

import matplotlib.pyplot as plt

# Data to plot
labels = ['Python', 'C++', 'Ruby', 'Java']
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = [0.1, 0, 0, 0]  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
         autopct='%.2f%%', shadow=True, startangle=90)

#plt.axis('equal')
plt.show()
```

# Subplot <

ศัพท์ที่ควรรู Figure กับ Axes

* Figure ภาพหลัก
* Axes แกนพิกัดของกราฟต่างๆ


มี Axes หลายๆ อัน อยู่ใน Figure เดียวกัน

plt.subplot(row, column, position)

อ่านเพิ่ม
 https://matplotlib.org/stable/gallery/subplots_axes_and_figures/subplots_demo.html

```
import matplotlib.pyplot as plt

plt.subplot(2, 3, 1)
plt.text(0.5, 0.5, '(2, 3, 1)', fontsize=14, ha='center')

plt.subplot(2, 3, 2)
plt.text(0.5, 0.5, '(2, 3, 2)', fontsize=14, ha='center')

plt.subplot(2, 3, 3)
plt.text(0.5, 0.5, '(2, 3, 3)', fontsize=14, ha='center')

plt.subplot(2, 3, 4)
plt.text(0.5, 0.5, '(2, 3, 4)', fontsize=14, ha='center')

plt.subplot(2, 3, 5)
plt.text(0.5, 0.5, '(2, 3, 5)', fontsize=14, ha='center')

plt.subplot(2, 3, 6)
plt.text(0.5, 0.5, '(2, 3, 6)', fontsize=14, ha='center')

plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np

x1 = np.linspace(0.1, 5.0)

plt.subplot(2, 1, 1)
plt.plot(x1, x1**2, 'o-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

plt.subplot(2, 1, 2)
plt.plot(x1, 1/x1, '.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.show()
```

ภายใน Figure กำหนดระยะห่างของ Axes ได้

อ่านเพิ่ม

https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.subplots_adjust.html

```
import matplotlib.pyplot as plt

fig = plt.figure()
fig.subplots_adjust(hspace=0.3, wspace=0.5)

plt.subplot(2, 3, 1)
plt.text(0.5, 0.5, '(2, 3, 1)', fontsize=14, ha='center')

plt.subplot(2, 3, 2)
plt.text(0.5, 0.5, '(2, 3, 2)', fontsize=14, ha='center')

plt.subplot(2, 3, 3)
plt.text(0.5, 0.5, '(2, 3, 3)', fontsize=14, ha='center')

plt.subplot(2, 3, 4)
plt.text(0.5, 0.5, '(2, 3, 4)', fontsize=14, ha='center')

plt.subplot(2, 3, 5)
plt.text(0.5, 0.5, '(2, 3, 5)', fontsize=14, ha='center')

plt.subplot(2, 3, 6)
plt.text(0.5, 0.5, '(2, 3, 6)', fontsize=14, ha='center')

plt.show()
```

แชร์แกน

```
import matplotlib.pyplot as plt

fig, ax = plt.subplots(2, 3, sharex='col', sharey='row')

ax[0,0].text(0.5, 0.5, '(2, 3, 1)', fontsize=14, ha='center')
ax[0,1].text(0.5, 0.5, '(2, 3, 2)', fontsize=14, ha='center')
ax[0,2].text(0.5, 0.5, '(2, 3, 3)', fontsize=14, ha='center')
ax[1,0].text(0.5, 0.5, '(2, 3, 4)', fontsize=14, ha='center')
ax[1,1].text(0.5, 0.5, '(2, 3, 5)', fontsize=14, ha='center')
ax[1,2].text(0.5, 0.5, '(2, 3, 6)', fontsize=14, ha='center')

plt.show()
```

ใช้ grid จัด loyout ของแต่ละ Axes

```
import matplotlib.pyplot as plt
grid = plt.GridSpec(2, 3, wspace=0.4, hspace=0.3)

plt.subplot(grid[0,0])
plt.subplot(grid[0,1:3])
plt.subplot(grid[1,0:2])
plt.subplot(grid[1,2])

plt.show()
```

อ้างอิง Axes ย่อยๆ ภายใน Figure โดย
มีเมธอดที่คล้ายตอนเรียกผ่าน plt.

* plt.xlabel() -> ax.set_xlabel()
* plt.ylabel() -> ax.set_ylabel()
* plt.xlim() -> ax.set_xlim()
* plt.ylim() -> ax.set_ylim()
* plt.title() -> ax.set_title()




```
import matplotlib.pyplot as plt

fig = plt.figure()
fig.subplots_adjust(hspace=0.4, wspace=0.9)

ax1 = fig.add_subplot(2, 3, 1)
ax1.text(0.5, 0.5, '(2, 3, 1)', fontsize=14, ha='center')

ax2 = fig.add_subplot(2, 3, 2)
ax2.text(0.5, 0.5, '(2, 3, 2)', fontsize=14, ha='center')

ax3 = fig.add_subplot(2, 3, 3)
ax3.text(0.5, 0.5, '(2, 3, 3)', fontsize=14, ha='center')

ax4 = fig.add_subplot(2, 3, 4)
ax4.text(0.5, 0.5, '(2, 3, 4)', fontsize=14, ha='center')

ax5 = fig.add_subplot(2, 3, 5)
ax5.text(0.5, 0.5, '(2, 3, 5)', fontsize=14, ha='center')

ax6 = fig.add_subplot(2, 3, 6)
ax6.text(0.5, 0.5, '(2, 3, 6)', fontsize=14, ha='center')

plt.show()
```

## เขียนประโยควนลูป ใช้ axis อ้างแต่ละกราฟได้

```
import matplotlib.pyplot as plt
import pandas as pd

fig = plt.figure()
fig.subplots_adjust(hspace=0.5, wspace=0.5)

# เก็บ axes ใน list
ax = [_ for i in range(4)]
for i in range(1,5):
  ax[i-1] = fig.add_subplot(2, 2, i)


df = pd.read_csv('https://www.w3schools.com/python/pandas/data.csv')

ax[0].plot(df.Duration)
ax[1].plot(df.Pulse)
ax[2].scatter(df.index, df.Maxpulse)
ax[3].scatter(df.index, df.Calories)

plt.show()
```

ใช้ add_axes เพิ่ม axes เข้าไปได้ อย่างฟรีสไตล์

fig.add_axes([left, bottom, width, height])



```
import matplotlib.pyplot as plt

fig = plt.figure()
ax1 = fig.add_axes([0.0, 0.6, 1, 0.4], ylim=(-1.2, 1.2))
ax2 = fig.add_axes([0.1, 0.1, 0.8, 0.4], ylim=(-1.2, 1.2))

x = np.linspace(0, 10, 100)
ax1.plot(np.sin(x))
ax2.plot(np.cos(x))

plt.show()
```

พล็อตเส้นในกราฟ 3 มิติ

```
import matplotlib.pyplot as plt
import numpy as np

#fig = plt.figure()
ax = plt.axes(projection='3d')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
#ax.set_zlabel('z-axis', rotation=180)

zline = np.linspace(0, 15, 100)
xline = np.sin(zline)
yline = np.cos(zline)

ax.plot3D(xline, yline, zline, 'gray')


plt.show()
```

พล็อตเส้น และ scatter ในกราฟ 3 มิติ

```
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')


zline = np.linspace(0, 15, 100)
xline = np.sin(zline)
yline = np.cos(zline)

ax.plot3D(xline, yline, zline, 'gray')

zdata = np.linspace(0, 15, 50)  + np.random.random(50)
xdata = np.sin(zdata) + 0.1*np.random.rand(50)
ydata = np.cos(zdata) + 0.1*np.random.rand(50)
ax.scatter3D(xdata, ydata, zdata)

plt.show()
```
